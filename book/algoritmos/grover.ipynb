# Algoritmo de Grover

O algoritmo de Grover {cite}`paper:grover` resolve o problema da busca em uma base de dados não estruturada. Esse é um algoritmo quantica que traz vatagem em ralação a sua contraparte classica. Enquanto computadores tradicionais tomam tempo $O(N)$ para encontrar um elemento um uma lista de $N$ elementos, o algoritmo de Grover consegue reduzir esse tempo para $O(\sqrt(N))$. Em ambos os casos, a complixidade é considera a melhor possivel para cada modelo de computação.

Como o algoritmo de Grover tem uma complexidade computacional menor que o algoromo classicol, independente do qual eficiente seja o computado classico, sempre vai exeistir uma entrada grade o suficiente onde o algoritmo quantico vai ser mais rapido.



```python
entrada = list(range(1, 1_000))
classical_speedup = np.linspace(10, 30, 10)

px.line(
    {
        "Tempo": np.array(
            [
                [*np.array(entrada), *np.array(list(map(np.sqrt, entrada))) * speedup]
                for speedup in classical_speedup
            ]
        ).flatten(),
        "Tamanho da Entrada": np.array(
            [[*entrada, *entrada] for _ in classical_speedup]
        ).flatten(),
        "Modelo de Computação": np.array(
            [
                [
                    *["Clássico"] * len(entrada),
                    *["Quântico"] * len(entrada),
                ]
                for _ in classical_speedup
            ]
        ).flatten(),
        "Speedup Clássico": np.array(
            [
                [*[speedup] * len(entrada), *[speedup] * len(entrada)]
                for speedup in classical_speedup
            ]
        ).flatten(),
    },
    x="Tamanho da Entrada",
    y="Tempo",
    color="Modelo de Computação",
    animation_frame="Speedup Clássico",
    title=f"Tempo de Computação Clássico vs. Quântico.",
)
```
